#Question 5:

 Given:

```python
sales = [12000, 15000, 11000, 18000, 17000]
```

Using a loop:
1. Calculate total revenue.
2. Count how many months had sales above 15000.
3. Find maximum sales without using built-in max().

```python
# 1. calculate total revenue

total_revenue = 0
for i in sales:
  total_revenue = total_revenue + i

print(total_revenue)


```

```python
# 2. Count how many months had sales above 15000
count =0
for i in sales:
  if i > 15000:
    count = count + 1

print(count)
```

```python
# 3. Find maximum sales without using built-in max().

max_sales = sales[0]

for i in sales:
  if i > max_sales:
    max_sales = i
print(max_sales)

```

#Question 9:
Write a generator that yields squares of numbers from 1 to 1,000,000.

Explain difference between:

● List comprehension

● Generator expression

Why are generators preferred in big data scenarios?

```python
def square_generator():
    for i in range(1, 1000001):
        yield i * i
gen = square_generator()

print(next(gen))
print(next(gen))
print(next(gen))
```

Explain difference between:

● List comprehension

● Generator expression

List Comprehension

* Creates the entire list immediately
* Stores all values in memory

Generator Comprehension

* Generates values one by one
* Does not store everything in memory

### 3. Why generators are preferred in Big Data scenarios

Big Data systems process millions of rows , streaming data

If all data is stored at once memory usage becomes huge system may slow down or crash Generators solve this problem.

#Question 10:
 Design a class:
class Product:
With:
● Attributes: name, price, quantity
● Method: calculate_revenue()
Create 3 objects and print their revenue.

```python
class product():
  def __init__(self, name, price, quantity):
    self.name = name
    self.price = price
    self.quantity = quantity

  def calculate_revenue(self):
    return self.price * self.quantity

a = product("Ram", 5000, 3)
b = product("Raj", 3500, 2)
c = product("Rai", 6000, 2)

print(a.calculate_revenue())
print(b.calculate_revenue())
print(c.calculate_revenue())
```
