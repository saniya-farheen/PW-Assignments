#Question 2:

You are given a CSV file sales.csv containing:

Tasks:
1. Read the file using Python.
2. Compute total revenue per product.
3. Write results to a new file revenue.csv.
4. Explain limitations of manual CSV handling vs Pandas.

```python
# 1. Read the file using Python

import csv

with open("/content/Sales.csv", "r") as file:
  reader = csv.DictReader(file)
  for row in reader:
    print(row)


```

```python
# 2. Compute Total revenue per product

Total_revenu= []

with open("/content/Sales.csv", "r") as file:
  reader = csv.DictReader(file)
  revenue_data = []
  for row in reader:
    product = row["product"]
    Prices = int(row["Prices"])
    Quantity = int(row["Quantity"])
    revenue = Prices * Quantity
    revenue_data.append([product, revenue])

revenue_data

```

```python
# 3.Write results to a new file revenue.csv.

with open("revenue.csv", "w", newline="") as file:
    writer = csv.writer(file)

    # Header
    writer.writerow(["product", "revenue"])

    # Data rows
    writer.writerows(revenue_data)

print("Revenue file created successfully!")
```

###4.Explain limitations of manual CSV handling vs Pandas

* Pandas handles datatypes automatically meanwhile manual csv needs type conversion

* Manual CSV is Harder for large dataset meanwhile Pandas is usually built for large dataset

* there is no built in filtering/grouping methods available for manual csv

* slower development in manual csv meanwhile in pandas have faster and cleaner workflow

#Question 3:
You receive JSON data from an API:


Tasks:
1. Load the JSON.
2. Extract all customer names.
3. Compute total purchase value.
4. Explain why JSON is widely used in ML pipelines.

```python
# 1. Load the JSON
import json

data = '''
{
    "customers": [
        {"id": 1, "name": "Amit", "purchase": 5000},
        {"id": 2, "name": "Sara", "purchase": 7000}
    ]
}
'''

json_data = json.loads(data)

print(json_data)
```

```python
# 2. Extract all customers name

customer_names = []

for customer in json_data["customers"]:
    customer_names.append(customer["name"])

print(customer_names)
```

```python
# 3. Compute total purchase value

total_purchase = 0

for customer in json_data["customers"]:
    total_purchase += customer["purchase"]

print("Total Purchase Value:", total_purchase)
```

### 4. Explain why JSON is widely used in ML pipelines.

Because JSON is easy for developers to read and understand

API's, WEb Applications, Databases and ML system commonly exchange data using json

Machine learning data often containes nested structures , JSON is ususally for nested data

```python
# 1. Create interactive bar chart of sales.

import plotly.express as px
import pandas as pd

# Sales data
df = pd.DataFrame({
    "Product": ["Laptop", "phone", "Tablet"],
    "Revenue": [250000, 200000, 105000]
})

# Interactive bar chart
fig = px.bar(
    df,
    x="Product",
    y="Revenue",
    title="Product Sales Revenue",
    hover_data=["Revenue"]
)

fig.show()
```

```python
# 2. Create pie chart for category distribution.
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    "Product": ["Laptop", "phone", "Tablet"],
    "Revenue": [250000, 200000, 105000]
})

fig = px.pie(
    df,
    names="Product",
    values="Revenue",
    title="Revenue Distribution by Product",


    # 3. Add hover information.
    hover_data=["Revenue"]
)

fig.show()
```
