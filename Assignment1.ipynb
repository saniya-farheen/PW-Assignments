##Question 7

You are given a customer review:
review = "The product quality is Excellent but delivery was Late"
Write Python code to:
1. Convert to lowercase
2. Count number of words
3. Check if the word "excellent" exists
4. Replace "Late" with "delayed"

```python

review = "The product quality is Excellent but delivery was Late"

#convert to lower case
print(review.lower())

# count the number of words
print(len(review))

# Check if the word "excellent" exists
print(review.find("Excellent"))

#Replace "Late" with "delayed"

print(review.replace("Late", "delayed"))
```

##Question 8

Given sales data:
sales = [12000, 15000, 11000, 18000, 17000]
1. Calculate total revenue.
2. Find average revenue.
3. Identify highest sales month.
4. Append new month sale = 20000.
5. Sort the list in descending order.



```python
sales = [12000, 15000, 11000, 18000, 17000]

#Calculate total revenue.
print(sum(sales))

# Find average revenue.
print(sum(sales)/len(sales))

#Identify highest sales month.
print(max(sales))

#Append new month sale = 20000.
sales.append(20000)
print(sales)

#Sort the list in descending order.
sales.sort(reverse= True )
print(sales)

```

##Question 9

You are building a simple sales analysis system.
Given:
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]
Tasks:
1. Compute total revenue for each product.
2. Identify which product generated maximum revenue.
3. Store revenues in a new list.
4. Print a formatted summary like:

```python
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

#compute total revenue of each product
revenue=[]

for i in range(len(prices)):

  revenue.append(prices[i] * quantities[i])

print(revenue)

#Identify which product generated maximum revenue.
print("product with maximum revnue", max(revenue))

```
